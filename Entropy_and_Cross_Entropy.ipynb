**Entropy**, as it relates to machine learning, is a measure of the randomness in the information being processed. The higher the entropy, the harder it is to draw any conclusions from that information. Flipping a coin is an example of an action that provides information that is random.

![](https://i.makeagif.com/media/12-19-2020/74a_3U.gif)

Cross Entropy loss or log loss is used to measure the performance of classification models whose output has a probability value between 0 and 1. Since there is more than one output in Neural Networks, we need to process the results. Argmax's derivative is bad, so backpropagation cannot be done. Therefore, activation functions such as Softmax are generally used (after leaving softmax, the sum of all outputs becomes 1.).

On the other hand, Cross Entropy is to get the result for each class after exiting the activation function to the negative log ```(-log(x) = log(1/x)) ```. After calculating this for each class, we add up the values ​​to find the "Total Cross Entropy". Our aim is to minimize this value as in every loss function.

In the weather example, logging the probabilities according to 2 gives us the information about how many bits are needed. Uncertanity Reduction is the opposite of the probability of the event. After taking the negative log, we multiply each value by its probability of occurrence and perform a weighted average. So entropy actually measures the average amount of information we get when you learn the weather each day.

The reason why Cross Entropy is preferred is that it can show the loss value in an unlimited range, not between 0-1. It also enables us to take big steps in backpropagation when our loss value is too large, and enables us to take small steps in the opposite scenario.

Let's think about a football game. There are 3 possibilities. Let's assume that the team we support has a winning rate of 0.7, a draw probability of 0.2, and a losing probability of 0.1:

````python
from math import log2
from matplotlib import pyplot
import matplotlib

matplotlib.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
fig, axes = pyplot.subplots(figsize=(12,12))
pyplot.style.use('fivethirtyeight')

# define distributions
events = ['win', 'draw', 'lose']
p = [0.33333, 0.33333, 0.33334]
q = [0.7, 0.2, 0.1]
print('P=%.3f Q=%.3f' % (sum(p), sum(q)))

# plot first distribution
pyplot.subplot(2,1,1)
pyplot.bar(events, p)
pyplot.title("Normalde Beklenen Dağılım", loc='center')

# plot second distribution
pyplot.subplot(2,1,2)
pyplot.bar(events, q)
pyplot.title("Eldeki Veri ile Beklenen Dağılım", loc='center')

# show the plot
pyplot.show()
````

````python
def cross_entropy(p, q):
	return -sum([p[i]*log2(q[i]) for i in range(len(p))])
````

````python
print(f"The resulting entropy value : {cross_entropy(p,q)}")
````
